```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
from sklearn.preprocessing import StandardScaler 
```

```python
import seaborn as sns 
sns.set() 
from sklearn.cluster import KMeans 
```

```python
con_data=pd.read_csv("Countries.csv")
```

```python
con_data.head()
```

```python
#Plotting the data 

plt.scatter(con_data['Longitude'], con_data['Latitude']) 
```

```python
cluster_data=con_data.iloc[:,1:3]
```

```python
ss = StandardScaler() 
X = pd.DataFrame(ss.fit_transform(cluster_data), columns=['Longitude', 'Latitude'])
```

```python
#Clustering

kmeans = KMeans(4) 
```

```python
kmeans.fit(cluster_data) 
```

```python
identified_clusters = kmeans.fit_predict(cluster_data) 
```

```python
new_data=con_data.copy()
```

```python
new_data['cluster_no']=identified_clusters
```

```python
plt.scatter(new_data['Longitude'],new_data['Latitude'],c=new_data['cluster_no'])
```
